### Random Forest:

Random Forest has a variety of applications, such as recommendation engines, image classifier and feature selection. It can be used to classify loyal loan applicants, identify fradulant activity and predict diseases. It lies at base of Boruta algorithm, which selects important features in a dataset.

### Steps:
1. **Dataset Splitting**:
   - The dataset (\(X\)) is randomly sampled multiple times to create subsets for training each tree. 
   - Each tree uses a random subset of features (\(N_1, N_2, \dots\)) for splitting decisions, ensuring diversity among the trees.

2. **Building Decision Trees**:
   - Multiple decision trees (Tree #1, Tree #2, Tree #3, etc.) are trained on different subsets of data.
   - Each tree independently predicts a class label (e.g., Class C, Class D, Class B, etc.).

3. **Majority Voting**:
   - For classification, the Random Forest aggregates the predictions from all the trees.
   - The class with the most votes becomes the **final class prediction**.

### Key Features of Random Forest:
- **Bagging**: Combines results from individual models to reduce variance and improve accuracy.
- **Randomness**: Introduces randomness in both data sampling and feature selection to make the model robust.
- **Majority Voting**: Ensures the final prediction reflects the consensus of all decision trees, minimizing the impact of any single tree's errors.

This approach makes Random Forest highly accurate and less prone to overfitting compared to a single decision tree.

### Decision Tree vs Random Forest:

| **Feature**            | **Decision Tree (DT)**                               | **Random Forest (RF)**                                  |
|:-------------------------|:------------------------------------------------------|:---------------------------------------------------------|
| **Definition**          | A single tree structure that splits data based on features. | An ensemble of multiple decision trees.                |
| **Complexity**          | Simple and interpretable.                           | More complex due to multiple trees.                    |
| **Accuracy**            | Lower accuracy, prone to overfitting.               | Higher accuracy, less prone to overfitting.            |
| **Stability**           | Sensitive to data changes (high variance).          | Stable due to averaging across trees.                  |
| **Overfitting**         | Likely to overfit on noisy datasets.                | Reduces overfitting through bagging and averaging.     |
| **Training Process**    | Faster as only one tree is built.                   | Slower as multiple trees are trained.                  |
| **Prediction Process**  | Faster since only one tree is used.                 | Slower due to combining predictions from all trees.     |
| **Feature Selection**   | Considers all features while splitting.             | Selects random subsets of features for diversity.       |
| **Use Case**            | Good for small, simple problems with clear splits.  | Suitable for complex, high-dimensional datasets.        |

**Key Takeaway:**
- Decision Tree is a foundational algorithm and works well for small datasets or when interpretability is crucial.
- Random Forest is a more powerful and robust ensemble method, offering better performance and generalization by combining multiple decision trees.
