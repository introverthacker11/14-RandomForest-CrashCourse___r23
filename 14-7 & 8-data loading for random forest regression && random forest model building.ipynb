```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

california = fetch_california_housing()
cal = pd.DataFrame(data = california.data, columns = california.feature_names)
cal['MedHouseVal'] = california.target
b = cal.head(200)
b.to_csv('california200.csv', index = False)
```

```python
b
```

```python
x = b.drop(['MedHouseVal'], axis = 1)
y = b['MedHouseVal']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 0, test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfreg = RandomForestRegressor()
rfreg.fit(x_train, y_train)
```

```python
y_pred = rfreg.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
from sklearn.metrics import r2_score
import numpy as np

print(f'R2:{r2_score(y_test, rfreg.predict(x_test))}')
print(f'Mean Absolute Error(MAE): {mean_absolute_error(y_test, rfreg.predict(x_test))}')
print(f'Mean Squared Error(MSE): {mean_squared_error(y_test, rfreg.predict(x_test))}')
print(f'Root Mean Squared Error: {np.sqrt(mean_absolute_error(y_test, rfreg.predict(x_test)))}')
```
